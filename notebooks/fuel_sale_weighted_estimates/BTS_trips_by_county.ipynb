### Calculate per capita distance traveled per county per month

```python
# NOTE: here we're dropping the trips > 500
multiplication_dict = {'Number of Trips <1' : 0.5, 'Number of Trips 1-3' : 2,
                       'Number of Trips 3-5' : 4,
       'Number of Trips 5-10' : 7, 'Number of Trips 10-25' : 13,
       'Number of Trips 25-50' : 37, 'Number of Trips 50-100' : 75,
       'Number of Trips 100-250' : 175, 'Number of Trips 250-500' : 375,
       'Number of Trips >=500' : 0}

county_merged["month"] = county_merged.Date.dt.month_name()

for key in multiplication_dict:
    county_merged[key] *= multiplication_dict[key]

county_merged.head()
```

```python
grouped = county_merged.groupby(["County FIPS", "Month", "total_pop"])[list(multiplication_dict.keys())].sum()

df = pd.DataFrame(grouped.sum(axis=1))
df = df.reset_index().rename(columns={0: "total_miles"})
df['per_capita'] = df['total_miles'] / df['total_pop']
df['County FIPS'] = df['County FIPS'].astype(str).str.zfill(5)

df.head()
```

```python
df.sort_values(by='per_capita', ascending=True).head(20)
```

```python
total_dist_by_county = df.query("per_capita > 0").groupby("County FIPS")[["total_miles"]].sum().reset_index()
```

```python
total_dist_by_county["STATE"] = total_dist_by_county["County FIPS"].str.slice(0,2)
total_dist_by_state = total_dist_by_county.groupby("STATE")["total_miles"].sum().reset_index().rename({"total_miles" : "total_miles_state"}, axis = 1)
write_to_ts3(total_dist_by_state, "BTS Trips/state_trip_distances.csv")
```

## Convert county aggregations to city

given a city geometry: sum over each county that overlaps with the city: find the overlap area percentage of the county, multiplied by the county per capita number of trips

```python
county_pops = county_gdf[["STATE", "COUNTY", "total_pop"]].rename({"total_pop" : "total_pop_county"}, axis = 1)
county_pops["County FIPS"] = county_pops["STATE"] + county_pops["COUNTY"] 
```

```python
CENSUS_VARS = {"B01001_001E" : "total_pop"}
national_tracts = ced.download("acs/acs5", 2020, CENSUS_VARS, state=ALL_STATES_AND_DC, county = "*",tract = "*", with_geometry=True).rename(CENSUS_VARS, axis = 1).to_crs(4326)
```

```python
dists_merged = national_tracts.merge(county_pops).merge(total_dist_by_county)
dists_merged["pop_weight"] = dists_merged["total_pop"] / dists_merged["total_pop_county"]
dists_merged["weighted_tract_miles"] = dists_merged["pop_weight"] * dists_merged["total_miles"]
```
